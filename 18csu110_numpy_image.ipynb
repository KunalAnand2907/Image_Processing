```python
import numpy as np
import matplotlib.pyplot as plt
```

<B>Loading an Image as a Numpy array</B>

```python
# By Matplotlib first read and then show it using imshow
A=plt.imread('earth.jpg') # To read the file
B=plt.imread('elon_musk.jpg')

plt.imshow(A)
plt.show()
plt.imshow(B)
plt.show()
# H = 0-772
# W = 0-800
```

```python
print(np.shape(A)) # or can use A.shape
print(type(A))
print(A.dtype)
print(np.shape(B))
```

```python
print(A) # To show in matrix form
print(B)
```

so our color image (192X263X3) is imported as an N-dimenstional array object available from numpy.

## Printing Images For different Channels

```python
A_red = A[:, :, 0]# in matplotlib h,w,channel
A_green = A[:, :, 1]
A_blue = A[:, :, 2]
print(A_red)
plt.figure()
plt.imshow(A_red) #For a single channel / grayscale image you need to mention the colourmap
plt.title('The RED channel')

plt.figure()
plt.imshow(A_green, cmap='gray')# Cmap to tell if it is grayscale or of rgb format
plt.title('The GREEN channel')

plt.figure()
plt.imshow(A_blue)
plt.title('The BLUE channel')

plt.show()
```

```python
help(plt.imshow)
```

```python
plt.imshow(A_green)
plt.axis('off')# to don't show the axis
plt.title('Green Image')
plt.show()
```

## create a random uint8 array of size i.e 2 d matrix of 400-Rx500-C with intensity numbers range from 0 to 255

```python
# random.randint- gives an array with random assign of numbers:numpy.random.randint(low, high=None, size=None, dtype=’l’)
x = np.random.randint(0, 256, (400, 500)).astype('uint8')
# Or can give like this (high,low,size,dtype='uint8')
print(x)
plt.imshow(x)
#plt.imshow(x, cmap='gray')
```

```python
x = np.random.randint(0, 256, (400,500,3)).astype('uint8')
# Or can give like this (high,low,size,dtype='uint8')
print(x)
```

# Lets try Python Imaging Library

```python
# Pillow library don;t store image as Array we need to convert them by .asarray()
from PIL import Image
pic=Image.open('earth.jpg')
pic
```

```python
# Format of image
type(pic)
```

```python
# Converting the image to array so as to preprocess the image
pic_array=np.asarray(pic)
print(pic_array.shape)
print(pic_array.size)
```

```python
# Show the image wrt axis
plt.imshow(pic_array)
```

```python
# Making a Copy of the earth picture
pic_copy=pic_array.copy()
```

## converting g,b to red
x,y,channel

```python
pic_copy[:,:,2]=0
pic_copy[:,:,1]=0
plt.imshow(pic_copy)
```

## converting r,b to green
x,y,channel

```python
pic_copy[:,:,0]=1
pic_copy[:,:,2]=1
plt.imshow(pic_copy)
```
