## FACE, EYE, SMILE DETECTION USING OPEN CV

```python
import numpy as np
import cv2
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_smile.xml')
nose_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'./cascade_files/haarcascade_mcs_nose.xml')
```

```python
#for color image
import cvlib as cv
img = cv2.imread("elon_musk.jpg",1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray,1.3,5)
cv2.putText(img, "webcam working: {}".format('yes'), (10, 20), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 5)
 #FACE DETECTION
for (x, y, w, h) in faces:
     cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
     roi_gray = gray[y:y+h, x:x+w]
     roi_color = img[y:y+h, x:x+w]
     cv2.putText(img, "face Detect: {}".format('yes'), (400, 50), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 3)
 #EYES DETECTION
     eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 18)
     for (ex, ey, ew, eh) in eyes:
         cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
         cv2.putText(img, "eye Detect: {}".format('yes'), (200, 100), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 3)
 #SMILE DETECTION
     smiles = smile_cascade.detectMultiScale(roi_gray, 1.7, 18)
     for (sx, sy, sw, sh) in smiles:
         cv2.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (0, 0, 255), 2)
         cv2.putText(img, "smile Detect: {}".format('yes'), (400, 150), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 3)
cv2.putText(img, "enter q to exit: {}".format('?'), (25, 150), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 5)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
